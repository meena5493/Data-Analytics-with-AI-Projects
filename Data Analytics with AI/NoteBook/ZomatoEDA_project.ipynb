```python
import pandas as pd
```

```python
df=pd.read_csv("zomato.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df=df.rename(columns={"rate (out of 5)":"Ratings"})
```

```python
df
```

```python
df=df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
df=df.rename(columns={"Unnamed: 0.1":"Order_ID"})
```

```python
df
```

```python
df["restaurant name"]=df["restaurant name"].str.replace(r"^'[@$#*]+","",regex=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
avg_rating=df["Ratings"].mean()
```

```python
df["Ratings"]=df["Ratings"].fillna(avg_rating)
```

```python
avg_cost=df["avg cost (two people)"].mean()
```

```python
df["avg cost (two people)"]=df["avg cost (two people)"].fillna(avg_cost)
```

```python
#What correlations exist among numeric features in the dataset
import seaborn as sns
numeric_col=["Ratings","num of ratings","avg cost (two people)"]
corr=df[numeric_col].corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#• Which restaurants are priced above the overall cost average
avg_price=df["avg cost (two people)"].mean()
print(avg_price)
df[df["avg cost (two people)"]>avg_price]
```

```python
#• Which restaurants have ratings above the dataset average
avg_rating=df["Ratings"].mean()
print(avg_rating)
df[df["Ratings"]>avg_rating]
```
