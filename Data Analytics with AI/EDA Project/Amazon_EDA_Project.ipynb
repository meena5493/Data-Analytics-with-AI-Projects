```python
#8. how much Revenue lost due to cancelled orders/Pending/Completed
cancelled_amount=df[df["Status"]=='Cancelled']["Total Sales"].sum()
print(cancelled_amount)
pending_amt=df[df["Status"]=='Pending']["Total Sales"].sum()
print(pending_amt)
completed_amt=df[df["Status"]=='Completed']["Total Sales"].sum()
print(completed_amt)
```
